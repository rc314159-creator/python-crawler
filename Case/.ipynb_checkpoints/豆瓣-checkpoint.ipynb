# è±†ç“£250ç¿»é¡µ

```python
url = 'https://movie.douban.com/top250'
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/123.0.0.0 Safari/537.36',
}
name = []
score = []
for i in range(10):
    para = {
        'start':25*i
    }
    res = requests.get(url, headers=headers, params=para, timeout=5)
    # éœ€è¦çš„æ•°æ®åœ¨æ ‡ç­¾ä¸­ï¼Œç”¨lxml
    html=etree.HTML(res.text)
    name.extend(html.xpath('//div[@class="hd"]/a/span[1]/text()'))
    score.extend(html.xpath("//div[@id='content']/div[@class='grid-16-8 clearfix']/div[@class='article']/ol[@class='grid_view']/li[*]/div[@class='item']/div[@class='info']/div[@class='bd']/div[@class='star']/span[@class='rating_num']/text()"))
# è£…åŒ…
list(zip(name,score))
```
